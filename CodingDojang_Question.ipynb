##**[Lv.1] 완전수 구하기**

```python
def divisor_sub_N(x) :    #1
    divisor = []
    for i in range(1,x) :
        if x % i == 0 :
            divisor.append(i)
    return divisor

N = int(input())    #2

result = []    #3

for i in range(1,N+1) :    #4
    if sum(divisor_sub_N(i)) == i :
        result.append(i)

print(result)
```

##**[Lv.2] 넥슨 입사문제 중에서**

```python
# My solution

generator_list = []    #1

for i in range(1, 5000) :    #2
    add_num = []
    for j in str(i) :    #3
        add_num.append(j)
    add_num.append(i)    #4
    generator_list.append(sum(map(int,add_num)))    #5

self_num = {x for x in range(1,5000)} - set(generator_list)    #6

print(sum(self_num))    #7
```

##**[Lv.2] 문자열 압축하기**

```python
# My solution
word = input()    #1

count = 1    #2
result = []

for i in range(len(word)) :    #3
    if i == len(word)-1 :    #4
        result.extend([word[i],count])
    else :
        if word[i] == word[i+1] :
            count += 1
        else :
            result.extend([word[i],count])
            count = 1

print(''.join(map(str,result))) #5
```

##**[Lv.2] 다음 입사문제 중에서**

```python
# My solution
n = list(map(int,input('입력하실 숫자를 띄어쓰기로 입력해주세요 >>> ').split()))    #1

for i in range(len(n)-1) :    #2
    sub_m = n[i+1] - n[i]
    if i == 0 :    #3
        sub = sub_m
        result = [(n[i],n[i+1])]
    elif sub_m < sub :     #4
        sub = sub_m
        result = [(n[i],n[i+1])]
    elif sub == sub_m :    #5
        result.append((n[i],n[i+1]))

for k in result :    #6
    print(k, end=' ')
```

##**[Lv.3] Spiral Array**

```python
n,m = map(int,input().split())    #1
print()

matrix = [[-1 for _ in range(m)] for _ in range(n)]    #2

x, y = 0, 0     #3
dx, dy = 0, 1
count = 0

while matrix[x][y] == -1 :    #4
    matrix[x][y] = count
    count += 1
    x, y = x+dx, y+dy
    if x == n or y == m or matrix[x][y] != -1 :     #5
        x, y = x-dx, y-dy
        dx, dy = dy, -dx
        x, y = x+dx, y+dy

for i in matrix :    #6 
    for j in i :
        print(f'{j:2d}', end=' ')
    print()    
```

```python
# 넘파이를 이용해서 풀이한 것
import numpy as np

n,m = map(int,input().split())    #1
print()

matrix = np.full((n,m),-1)  #2

x, y = 0, 0     #3
dx, dy = 0, 1
count = 0

while matrix[x,y] == -1 :    #4
    matrix[x,y] = count
    count += 1
    x, y = x+dx, y+dy
    if x == n or y == m or matrix[x,y] != -1 :     #5
        x, y = x-dx, y-dy
        dx, dy = dy, -dx
        x, y = x+dx, y+dy

for i in matrix :    #6 
    for j in i :
        print(f'{j:2d}', end=' ')
    print()  
```

```python
# 참고한 풀이

X, Y = map(int, input("행열 숫자를 입력하세요: ").split(','))

table = [[-1 for _ in range(Y)] for _ in range(X)]

x, y = 0, 0
dx, dy = 0, 1
count = 0

while table[x][y] == -1:
    table[x][y] = count
    count += 1
    x, y = x+dx, y+dy    
    if x == X or y == Y or table[x][y] !=-1:
        x, y = x-dx, y-dy
        dx, dy = dy, -dx
        x, y = x+dx, y+dy

for i in table:
    for j in i:
        print (f'{j:3d}', end=' ')
    print ()
```
